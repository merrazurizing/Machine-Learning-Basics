```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import style

#poner Estilo a los graficos
style.use('bmh')

#array de 10 numeros random
numbers=np.linspace(10,100,30)

#Numero de saltos del histrograma
jumps = [0,15,30,45,60,75,90,105]

plt.title("Histogram Test")
plt.xlabel("Xs")
plt.ylabel("Ys")

#Histograma input(numeros,saltos,tipo de histograma)
plt.hist(numbers,jumps,histtype='bar')
plt.show()
```

```python
plt.hist(numbers,jumps,histtype='step')
plt.show()
```
