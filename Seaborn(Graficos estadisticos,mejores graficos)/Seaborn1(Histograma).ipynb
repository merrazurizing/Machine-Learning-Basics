```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import scipy as sp
import seaborn as sb

#Importa base de datos
database = sb.load_dataset('diamonds')
print(database)
```

```python
#Grafico de Distribucion de la variable carat , es un histograma con la funcion en forma de curva encima
sb.displot(database['carat'])

#mostrar grafico
plt.show()
```

```python
#Bins es el tama√±o de el histograma
sb.displot(database['carat'],bins=1)

#mostrar grafico
plt.show()
```
